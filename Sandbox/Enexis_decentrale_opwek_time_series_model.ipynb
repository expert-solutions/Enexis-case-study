# Simple time series model

#### Enexis decentrale opwek data are prepared to combine with CBS Zonenstrom dataset

```python
import pandas as pd
import numpy as np
import sys
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
c_path = "./data/"
v_file = ("Enexis_decentrale_opwek_kv_(zon_pv)_01012020",                                    # 0
          "Enexis_decentrale_opwek_kv_(zon_pv)_01072020",                                    # 1
          "Enexis_decentrale_opwek_kv_(zon_pv)_01012021",                                    # 2
          "Enexis_decentrale_opwek_kv_(zon_pv)_01072021",                                    # 3
          "Enexis_decentrale_opwek_kv_(zon_pv)_01012022")                                    # 4
```

```python
df = pd.DataFrame()
for i in range(len(v_file)):
    df_new = pd.read_csv(filepath_or_buffer = c_path + v_file[i] + ".csv",
                         sep                = ';',
                         decimal            = ',',
                         thousands          = '.',
                         encoding           = 'UTF-8')
    df_new.columns = df_new.columns.str.strip()
    df_new['Year'] = v_file[i][-8:]
    df = pd.concat([df, df_new], axis = 0)
```

```python
df['Year'] = pd.to_datetime(df.Year, format = "%d%m%Y")
```

```python
df['Year'].value_counts()
```

```python
df = df[['Year', 'CBS Buurtcode', 'Aantal aansluitingen in CBS-buurt', 'Aantal aansluitingen met opwekinstallatie', 
         'Opgesteld vermogen']]
```

```python
df['CBS Buurtcode'].value_counts()
```

```python
df['CBS Buurtcode'] = df['CBS Buurtcode'].astype(str)
```

```python
def Formatbuurt(code):
    if 'BU' in code:
        return code
    elif len(code) == 10:
        return 'BU' + code[:-2]
    elif len(code) == 9:
        return 'BU' + '0' + code[:-2]
    elif len(code) == 8:
        return 'BU' + '00' + code[:-2]
    elif len(code) == 7:
        return 'BU' + '000' + code[:-2]
    else:
        return np.nan
```

```python
df['CBS Buurtcode'][2]
```

```python
df['CBS Buurtcode'] = df['CBS Buurtcode'].apply(Formatbuurt)
```
